```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from statsmodels.api import OLS
%matplotlib inline
```

```python
from length_hist import length
x = []
for key in length:
    x += [key] * length[key]

fig, ax = plt.subplots(1,1, figsize = (20,12))
ax.set_facecolor((1,1,.95))
ax.set_xlabel("Songs in Playlist", fontsize = 25, labelpad = 20)
ax.set_ylabel("Frequency", fontsize = 25, labelpad = 20)
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=20)
ax.grid(alpha = .1)
ax.set_title("Distribution of Playlist Length", fontsize = 30, pad = 50)
ax.set_xlim(0,260)
ax.set_xticks(np.round(np.linspace(0, 250, 11)))
ax = plt.hist(x, bins = 40, color = 'springgreen', alpha = .8)

plt.savefig("Length Hist", format = 'png')


```

```python
from followers import followers

x = []
for key in followers:
    x += [key] * followers[key]

fig, ax = plt.subplots(1,2, figsize = (30,12))
fig.suptitle("Histogram of Playlist Follower Count", fontsize = 30)
fig.tight_layout(pad = 12, w_pad = 22)
titles = ["Linear X Axis", "Logorithmic X Axis"]
ax[1].set_xscale('symlog', linthreshy = 10)
ax[0].set_xlim(5,3000)
bins = [1000,1000]
for col in range(2):
    ax[col].set_facecolor((1,1,.95))
    ax[col].set_xlabel("Follower Count", fontsize = 25, labelpad = 20)
    ax[col].set_ylabel("Frequency (log scale)", fontsize = 25, labelpad = 20)
    ax[col].xaxis.set_tick_params(labelsize=20)
    ax[col].yaxis.set_tick_params(labelsize=20)
    ax[col].grid(alpha = .1)
    ax[col].set_title(titles[col], fontsize = 25, pad = 30)
    ax[col].set_yscale('symlog', linthreshy = 10)
    ax[col].hist(x, bins = bins[col], color = 'springgreen', alpha = .8)


plt.savefig("Followers Hist.png", format = 'png')
```
